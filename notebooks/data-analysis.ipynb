```python
import os
import random
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.image import imread
import pandas as pd
```

```python
data_dir = '../data/seg_train/'
class_counts = {}
for class_name in os.listdir(data_dir):
    class_path = os.path.join(data_dir, class_name)
    if os.path.isdir(class_path):
        class_counts[class_name] = len(os.listdir(class_path))
```

```python
df = pd.DataFrame(list(class_counts.items()), columns=["Class", "Count"])

plt.figure(figsize=(10, 6))
sns.barplot(x="Class", y="Count", data=df)
plt.title("Image by Class Distribution")
plt.xlabel("Class")
plt.ylabel("Image count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('../results/plots/Imabe_by_Class_Distribution.jpg', format='jpg', dpi=300)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
i = 1
for class_name in df["Class"]:
    class_path = os.path.join(data_dir, class_name)
    img_name = random.choice(os.listdir(class_path))
    img_path = os.path.join(class_path, img_name)
    img = imread(img_path)
    plt.subplot(2, 3, i)
    plt.imshow(img)
    plt.title(class_name)
    plt.axis("off")
    i += 1
plt.suptitle("Examples of images")
plt.tight_layout()
plt.savefig('../results/plots/Examples_of_Images.jpg', format='jpg', dpi=300)
plt.show()
```
